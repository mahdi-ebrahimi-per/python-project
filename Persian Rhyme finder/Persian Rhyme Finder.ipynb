### dictionary (meaning)

```python


def search_syn_from_word(word):
    index = 0
    for i in words:
        i = i.replace(" ", "")
        if i == word:
            str_syn = syn[index]
            if "&" in str_syn:
                syn_scope = int( str(str_syn).find('&') )
                slicing = slice(1 , syn_scope - 1) 
                return str_syn[slicing]
            
            return str_syn[1:]

        index += 1



def search_opp_from_word(word):
    index = 0
    for i in words:
        i = i.replace(" ", "")
        if i == word:
            str_syn = syn[index]
            if "&" in str_syn:
                opp_scope = int(str(str_syn).find('&'))
                slicing = slice(opp_scope + 2, len(str_syn)) 
                return str_syn[slicing]
            
            else:
                return None


        index += 1


search = "آباجی"

print(search_syn_from_word(search))
print(search_opp_from_word(search))
```
